## 3. Make data compatible:  
### a. Convert labels to one-hot-vectors.
### b. Print the label for y_train[0].
### c. Split the dataset into training, testing, and validation set. (Hint: First split images and labels into training and testing set with test_size = 0.3. Then further split test data into test and validation set with test_size = 0.5)
### d. Check the shape of data, Reshape data into shapes compatible with Keras models if it’s not already. If it’s already in the compatible shape, then comment in the notebook that it’s already in compatible shape.

```python
le = LabelEncoder().fit(y) # fit the labelencoder
y = le.transform(y) # apply label encoder transformation

transformed_labels = le.transform(labels) # get the transformed labels index
l2i= dict(zip(labels, transformed_labels)) # create a label to transformed label mapping
i2l = dict(zip(transformed_labels, labels)) # create a transformed label to label mapping

y = k.utils.to_categorical(y) # apply the keras ohe
```

```python
print(y[0])
```

```python
# split first into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# then split into val and test
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=42)
```

## 4. Building CNN:  
### a. Define layers.
### b. Set optimizer and loss function. (Use Adam optimizer and categorical crossentropy.)

```python
tf.compat.v1.reset_default_graph()
model = k.models.Sequential()
model.add(Conv2D(filters=32, kernel_size=3, activation='relu', padding = 'same', input_shape = (128,128,3)))
model.add(BatchNormalization())
model.add(Conv2D(filters=32, kernel_size=3, activation='relu', padding = 'same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))


model.add(Conv2D(filters=64, kernel_size=3, activation='relu', padding = 'same'))
model.add(BatchNormalization())
model.add(Conv2D(filters=64, kernel_size=3, activation='relu', padding = 'same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))


model.add(Conv2D(filters=128, kernel_size=3, activation='relu', padding = 'same'))
model.add(BatchNormalization())
model.add(Conv2D(filters=128, kernel_size=3, activation='relu', padding = 'same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(filters=128, kernel_size=3, activation='relu', padding = 'same'))
model.add(BatchNormalization())
model.add(Conv2D(filters=128, kernel_size=3, activation='relu', padding = 'same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(filters=128, kernel_size=3, activation='relu', padding = 'same'))
model.add(BatchNormalization())
model.add(Conv2D(filters=128, kernel_size=3, activation='relu', padding = 'same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))


model.add(Flatten())
model.add(Dropout(0.3))
model.add(Dense(1024,activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(no_labels, activation='softmax'))
```

```python
model.summary()
```

```python
# compile the model using adam optimizer and categorical crossentriopy as loss function
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer = 'adam')
```

```python
# create a callback for saving best model seen so far
checkpoint_filepath = 'saved_models/model'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath,
    save_weights_only=False,
    monitor='val_accuracy',
    mode='max',
    save_best_only=True )

# create a callback for early stipping
model_early_sopping = tf.keras.callbacks.EarlyStopping (
    monitor='val_accuracy', min_delta=0, patience=10, verbose=0,
    mode='max', baseline=None, restore_best_weights=True
)
```

```python
# his = model.fit(x=X_train, y=y_train, batch_size=32, epochs=500, validation_data=(X_val, y_val), callbacks=[model_checkpoint_callback])
```

```python
# create a data generator object to create new data for training by transforming images
data_generator = tf.keras.preprocessing.image.ImageDataGenerator(
    featurewise_center=True,
    featurewise_std_normalization=True,
    rotation_range=20,
    vertical_flip=True,
    width_shift_range=0.1, 
    height_shift_range=0.1, 
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)
train_generaor = data_generator.flow(X_train, y_train, batch_size=32)
step_per_epoch = X_train.shape[0] // 32
```

## 5. Fit and evaluate model and print confusion matrix.  

```python
#call tje fit generator function to start training incorporating the image generation by transformation
model.fit_generator(train_generaor, validation_data=(X_val, y_val), callbacks=[model_checkpoint_callback] , steps_per_epoch=step_per_epoch, epochs = 1000)
```

```python
# load the best saved model 
m = tf.keras.models.load_model('saved_models/model/')
y_pred = m.predict(X_test) # predic on the test set
y_pred = [i.argmax() for i in y_pred] # get the label with max prop for predictions
y_act = [i.argmax() for i in y_test] # get the label with max prop (in this case 1) for actual y

print('accuracy on test data is',accuracy_score(y_act,y_pred)) # get the accuracy score
```

```python
# a method to build the confusion matrix
def draw_cm( actual, predicted ):
    ticks = [i for i in range(len(labels))]
    cm = confusion_matrix( actual, predicted )
    sns.heatmap(cm, annot=True,  fmt='.2f', xticklabels = ticks , yticklabels = ticks )
    plt.ylabel('Observed')
    plt.xlabel('Predicted')
    plt.show()
```

```python
plt.figure(figsize= (15,10))
draw_cm(y_act,y_pred) # plot the confusion matrix
```

## We see a very excellent result in the test set with accuracy about 97%

### We see a case where prediction is label 6 but actual is label 0. We see 6 of such cases and this is the only instance where the model is is very slightly. Lets dig in this case and see the images

```python
# a function to get the missclasifed indexs for a set of predicted and observed labels
def get_missclasifed_index(y_,y, obs, pred ): 
    index = []
    for i,j in enumerate(zip(y,y_)):
        if j[0] == obs and j[1] == pred:
            index.append(i)
    return index

ind = get_missclasifed_index(y_pred, y_act, 0, 6) # get indexs for pred 6 where actual was 0
other_index = [i[0] for i in enumerate(y_act) if i[1] == 6] # get indexes where actual is 6 for some comparison
```

```python
# Visualize the missclasifed cases we are looking
pos = 1
plt.figure(figsize= (15,10))  # setting the figure size
for i in ind[:12]: # loop through the unique labels
    plt.subplot(3, 4, pos) 
    plt.imshow(cv2.cvtColor(X_test[i], cv2.COLOR_BGR2RGB)) # plot the image
    plt.title(i)
    pos += 1  # to plot over the grid one by one
```

```python
# Visualize some example of the class which is incorrectly predicted
pos = 1
plt.figure(figsize= (15,10))  # setting the figure size
for i in other_index[:12]: # loop through the unique labels
    plt.subplot(3, 4, pos) 
    plt.imshow(cv2.cvtColor(X_test[i], cv2.COLOR_BGR2RGB)) # plot the image
    plt.title(i)
    pos += 1  # to plot over the grid one by one
```
