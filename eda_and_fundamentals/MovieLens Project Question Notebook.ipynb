### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use item dataset to get count of movies released in a particular year, store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.

```python
genre_columns = list(movie.columns)[4:-1] # get genre columns
agg_dic = {gen : 'sum' for gen in genre_columns} # make agg dict for sum agg
agg_dic['movie title'] = 'count' 

# groupby release year 
year_groupby = movie.groupby('release year').agg(agg_dic).rename({'movie title': 'total'}, axis = 1)
year_groupby_perc =  (year_groupby.div(year_groupby['total'], axis= 0) * 100) # take percent by dividing total movie in a genre by total movie
year_groupby_perc_ra_10 = year_groupby_perc.rolling(window=10).mean() # take a moving average to get a sence of trend

fig = plt.figure(figsize=(20,70))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

for i, col in enumerate(genre_columns):
    ax = fig.add_subplot(9, 2, i+1)
    g = year_groupby_perc[col].plot(ax = ax)
    year_groupby_perc_ra_10[col].plot(ax = ax)
    g.axes.set_title(f"Trend of Genre {col.upper()}")
    g.set_xlabel("Release Year" )
    g.set_ylabel("Percentage of Total Movies" )
    g.set_ylim(0,101)
    
```

### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
ratings_temp = ratings.copy() # make a copy of ratings df
ratings_temp['rating_count'] = 1 # add a dummy with all value 1 so that we can groupy by in single go
ratings_temp = ratings_temp.groupby('movie id', as_index = False).agg({
    'rating' : 'mean',
    'rating_count': 'count'
})

movie_temp = movie[['movie id','movie title']]
merge = movie_temp.merge(ratings_temp,on = 'movie id') # merge on the movie id
merge = merge[merge['rating_count'] > 100] # filter rating count more than 100
merge = merge.sort_values('rating', ascending = False).rename({'rating': 'average rating'}, axis = 1).reset_index()
merge[['movie title','average rating']].head(25)
```

3. you might want ot use the .sum(), .div() function here.

4. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python
# Marge the three tables
merge = movie.merge(ratings,on='movie id') 
merge = merge.merge(user, on = 'user id')

# seperate male and female data
male = merge[merge['gender'] == 'M']
female = merge[merge['gender'] == 'F']


# function to calculate who watches more movies of a particualr as a percentage of their apperenace
def find_who_watch_more(genre):
    pert_male = male[genre].sum() / len(male)*100  # divide the total rating of a genre  by total rating  for all genre
    pert_female = female[genre].sum() / len(female)*100
    print(f'{pert_male:.2f}% of Males watch movie of genre {genre}')
    print(f'{pert_female:.2f}% of Females watch movie of genre {genre}')
    if pert_male > pert_female:
        first = 'Male'
        second = 'Female'
    else:
        first = 'Female'
        second = 'Male'
    print(f'Therefore {first} watch more movies of genre {genre} than {second}')
    
find_who_watch_more('Drama')
print()
find_who_watch_more('Sci-Fi')
print()
find_who_watch_more('Romance')
```
